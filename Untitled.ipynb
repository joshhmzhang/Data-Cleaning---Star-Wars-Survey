## Clean the Data

* There is only one column contains float and lots of the columns contain more than 20% null values.

### Cleaning - First Column

```python
# There is one null in RespondentID. Since the ID should be unique, the null needs to be removed
star_wars=star_wars[star_wars['RespondentID'].notnull()]
star_wars['RespondentID'].isnull().sum()
```

```python
star_wars.head()
```

### Cleaning - Column 2-3

```python
#clean the columns 2-3: convert 'Yes'/'No' to True/False

for item in star_wars.columns[1:3]:
    star_wars[item]=star_wars[item].apply(lambda x: True if x=='Yes' else False)
            
star_wars.head()
```

### Cleaning - Column 4-9

```python
# find the content for column 4-9
have_seen_col=star_wars.columns[3:9]
for item in have_seen_col:
    print(star_wars[item].unique())
```

* The columns 4-9 indicating if the respondent have seen the star wars eqisode from I to VI. nan could mean no
* To make it more clear, we will need to rename the column to the episode, and update the content as True or False 

```python
# rename the columns
name_chane={'Which of the following Star Wars films have you seen? Please select all that apply.':'Seen I',
            'Unnamed: 4':'Seen II','Unnamed: 5':'Seen III','Unnamed: 6':'Seen IV','Unnamed: 7':'Seen V',
            'Unnamed: 8':'Seen VI'}
star_wars.rename(columns=name_chane,inplace=True)
star_wars.columns[3:9]
```

```python
#update the content for columns 4-9

for col in star_wars.columns[3:9]:
    star_wars[col]=star_wars[col].apply(lambda x: False if pd.isnull(x) else True)
star_wars.head()
```

### Cleaning - Column 9-15
* the ranking for episode I-VI

```python
star_wars.iloc[:,9:15].head(5)
```

```python
#rename the column
name_chane={'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'Rank I',
            'Unnamed: 10':'Rank II','Unnamed: 11':'Rank III','Unnamed: 12':'Rank IV','Unnamed: 13':'Rank V',
            'Unnamed: 14':'Rank VI'}
star_wars.rename(columns=name_chane,inplace=True)

#change data type to float
star_wars.iloc[:,9:15]=star_wars.iloc[:,9:15].astype(float)
star_wars.iloc[:,9:15].mean()
```

### Cleaning - Columns 15-29
* Regardinf the faviourate characters. But have no column names. So I remove these data 

```python
star_wars.iloc[:,15:30].head(2)
```

```python
star_wars=star_wars.drop(star_wars.columns[15:30],axis=1)
```

```python
star_wars.columns[15:19]
```

```python

for item in star_wars.columns[15:18]:
    star_wars[item]=star_wars[item].apply(lambda x: True if x=='Yes' else False)
    
star_wars.head()
```

### Which episode is more popular?  Which is the best?

```python
episode_pop=star_wars.iloc[:,3:9].sum()
episode_pop.plot.bar()
```

```python
episode_rank=star_wars.iloc[:,9:15].mean()
episode_rank.plot.bar()
```

It is clear that although the newer episode have been watched by more people, the original seems more popular

### Do male and female have a diverse taste?

```python
male=star_wars[star_wars['Gender']=='Male']
female=star_wars[star_wars['Gender']=='Female']
episode_pop_m=male.iloc[:,3:9].sum()
episode_pop_f=female.iloc[:,3:9].sum()

episode_rank_m=male.iloc[:,9:15].mean()
episode_rank_f=female.iloc[:,9:15].mean()
```

```python
fig,axarr=plt.subplots(2,2,figsize=(20,5))

plt.subplots_adjust(hspace=.8,wspace=.2)
episode_pop_m.plot.bar(ax=axarr[0][0])
ax=axarr[0][0].set_title('Episode watched Number by Male',fontsize=15)
episode_pop_f.plot.bar(ax=axarr[0][1],color='mediumvioletred')
ax=axarr[0][1].set_title('Episode watched Number by Female',fontsize=15)
episode_rank_m.plot.bar(ax=axarr[1][0])
ax=axarr[1][0].set_title('Episode ranking by Male',fontsize=15)
episode_rank_f.plot.bar(ax=axarr[1][1],color='mediumvioletred')
ax=axarr[1][1].set_title('Episode ranking by Female',fontsize=15)
```

More males watches episodes 1-3, but Episode III is the best since both Male and Female love it the most

## Any diference by Age groups?

```python
age_18=star_wars[star_wars['Age']=='18']
age_30=star_wars[star_wars['Age']=='30']
age_45=star_wars[star_wars['Age']=='45']
age_60=star_wars[star_wars['Age']=='60']

rank_18=age_18.iloc[:,9:15].mean()
rank_30=age_30.iloc[:,9:15].mean()
rank_45=age_45.iloc[:,9:15].mean()
rank_60=age_60.iloc[:,9:15].mean()
```

```python
fig,axarr=plt.subplots(2,2,figsize=(20,5))

plt.subplots_adjust(hspace=.8,wspace=.2)
rank_18.plot.bar(ax=axarr[0][0])
ax=axarr[0][0].set_title('Episode ranking by 18 yeas old',fontsize=15)
rank_30.plot.bar(ax=axarr[0][1])
ax=axarr[0][1].set_title('Episode ranking by 30 yeas old',fontsize=15)
rank_45.plot.bar(ax=axarr[1][0])
ax=axarr[1][0].set_title('Episode ranking by 45 yeas old',fontsize=15)
rank_60.plot.bar(ax=axarr[1][1])
ax=axarr[1][1].set_title('Episode ranking by 60 yeas old',fontsize=15)
```

Interestingly, young people aged around 30 are prefer the old episode while the episode III is clearly still the top choice for all groups 

### Which got the most fans:Star Wars film franchise? Expanded Universe? or The Star Trek franchise?
